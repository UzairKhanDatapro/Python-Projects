## Size

```python
# Count Vs Size
ax=sns.countplot(x='Size',data=df,hue='Size')
sns.set({'figure.figsize':(7,3)})

for bars in ax.containers:
 ax.bar_label(bars)
    
#No of Order count Vs Size with lables
```

```python
# Quantity Vs Size
S_Qty=df.groupby(['Size'] ,as_index=False)['Quantity'].sum().sort_values(by='Quantity',ascending=False)

sns.barplot(x='Size',y='Quantity',data=S_Qty)
sns.set({'figure.figsize':(7,3)})
```

### Category

```python
df['Category'] = df['Category'].astype(str)
column_data = df['Category']
plt.hist(column_data, bins=30, edgecolor='Black')
plt.xticks(rotation=0)
plt.show()
sns.set({'figure.figsize':(7,3)})
```

##### Note: From above Graph you can see that most of the  buyers are buying T-shirts 

```python
# Checking B2B Data  by using pie chart 
B2B_Check = df['B2B'].value_counts()

#  Plot the pie chart
plt.pie(B2B_Check, labels=B2B_Check, autopct='%1.1f%%')
#plt.axis('equal')
plt.show()
```

```python
# Checking B2B Data  by using pie chart 
B2B_Check = df['B2B'].value_counts()

#  Plot the pie chart
plt.pie(B2B_Check, labels=B2B_Check.index, autopct='%1.1f%%')
#plt.axis('equal')
plt.show()
```

##### Note : From above chart  we can see that maximum i.e. 99.3% of buyers are retailers and 0.7% are B2B buyers

```python
#  Prepare data for pie chart
a1=df['Fulfilment'].value_counts()
# Step 4: Plot the pie chart
fig, ax=plt.subplots()

ax.pie(a1,labels=a1.index, autopct='%1.1f%%', radius=0.7, wedgeprops=dict(width=0.6))
ax.set(aspect='equal')

plt.show()
```

##### Note: From above chart you can see that most of the  Fulfilment  are  amazon 

```python
# Prepare data for scatter plot
x_data = df['Category']  
y_data = df['Size'] 

# Plot the scatter plot
plt.scatter(x_data, y_data)
plt.xlabel('Category')  
plt.ylabel('Size')  
plt.title('Scatter Plot') 
plt.show()
sns.set({'figure.figsize':(7,3)})
```

### Ship State

```python
# Plot count of cities by state
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

```python
# top_10_States 
top_10_state = df['ship-state'].value_counts().head(10)
# Plot count of cities by state
plt.figure(figsize=(12, 6))
sns.countplot(data=df[df['ship-state'].isin(top_10_state.index)], x='ship-state',hue='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of  State')
plt.xticks(rotation=45)
plt.show()
```

#### Note: From above Graph you can see that most of the  buyers are Maharashtra state
